# Classes OOP

```python
class Person():
    job = ""

    # initializer method
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    # method
    def print_name(self):
        print(self.name)
```

```python
person1 = Person("Ufuk", 47, "Male")
```

```python
person1.name
```

```python
person1.print_name()
```

```python
class Dog():

    year = 7

    def __init__(self, age):
        self.age = age
        # first option
        self.dog_human_age = age * self.year

    # second option
    def human_age(self):
        return self.age * Dog.year # Dog.year -> self.year
```

```python
my_dog = Dog(5)
```

```python
my_dog.age
```

```python
my_dog.dog_human_age
```

```python
my_dog.human_age()
```

## Inheritance

```python
class Musician():
    def __init__(self, name):
        self.name = name
        print("musician class")

    def test1(self):
        print("test1")

    def test2(self):
        print("test2")
```

```python
class Musician_plus(Musician):

    def __init__(self, name):
        Musician.__init__(self, name)
        print("musician plus")

    def test3(self):
        print("test3")
```

```python
musician1 = Musician("Safiye")
```

```python
musician2 = Musician_plus("Can")
```

## Polymorphism

```python
class Banana():

    def __init__(self, name):
        self.name = name

    def info(self):
        return f"100 calorie {self.name}"

class Apple():

    def __init__(self, name):
        self.name = name

    def info(self):
        return f"150 calorie {self.name}"
```

```python
banana = Banana("banana")
```

```python
apple = Apple("apple")
```

```python
banana.info()
```

```python
apple.info()
```

```python
fruit_new_list = [banana, apple]
```

```python
for fruit in fruit_new_list:
    print(fruit.info())
```

## Encapsulation

```python
class Phone():

    def __init__(self, name, price):
        self.name = name
        self.__price = price

    def info(self):
        print(f"{self.name} price is: {self.__price}")

    def changePrice(self, price):
        self.__price = price
```

```python
iphone = Phone("Iphone 17", 1000)
```

```python
iphone.info()
```

```python
iphone.changePrice(300)
```

```python
iphone.info()
```

## Abstraction

```python
from abc import ABC, abstractmethod
```

```python
class Car(ABC):

    @abstractmethod
    def max_speed(self):
        pass
```

```python
class Togg(Car):

    def max_speed(self):
        print("200km")
```

```python
togg = Togg()
```

```python
togg.max_speed()
```
