```python
# merge
```

```python
df_merge1 = pd.read_csv('00_files/07_merge_data1.csv')
```

```python
df_merge1
```

```python
df_merge2 = pd.read_csv('00_files/07_merge_data2.csv')
```

```python
df_merge2
```

```python
# merge - outer join
```

```python
df_merged_outer = pd.merge(df_merge1, df_merge2, on = "Employee_ID", how="outer")
```

```python
df_merged_outer
```

```python
# merge - left join
```

```python
df_merged_left = pd.merge(df_merge1, df_merge2, on ="Employee_ID", how="left")
```

```python
df_merged_left
```

```python
# merge - right join
```

```python
df_merged_right = pd.merge(df_merge1, df_merge2, on ="Employee_ID", how="right")
```

```python
df_merged_right
```

```python
# merge - inner join
```

```python
df_merged = pd.merge(df_merge1, df_merge2, on="Employee_ID", how="inner")
```

```python
df_merged
```
