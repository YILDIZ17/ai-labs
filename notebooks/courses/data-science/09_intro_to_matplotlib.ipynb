```python
age_list = [10,20,30,30,30,40,50,60,70,75]
```

```python
weight_list = [20,60,80,85,86,87,70,90,95,90]
```

```python
plt.plot(age_list, weight_list, "y") # y -> yellow, g, b, r
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age VS Weight")
plt.savefig("00_files/09_test_savefig_cmd")
plt.show()
```

```python
numpy_array1 = np.linspace(0,10,20)
```

```python
numpy_array2 = numpy_array1 **3
```

```python
plt.plot(numpy_array1, numpy_array2, "b*-") # *, -, *-, --
plt.show()
```

```python
plt.subplot(1,2,1) # row, column, number of graphic
plt.plot(numpy_array1, numpy_array2, "g*-")
plt.subplot(1,2,2)
plt.plot(numpy_array2, numpy_array1, "r--")
plt.show()
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.3,0.3])
figure_axes.plot(numpy_array1, numpy_array2, "y")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("Graph Title")
plt.show()
```

```python
new_fig = plt.figure(dpi=90)
new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])
new_axes.plot(numpy_array1, numpy_array1 ** 2, "r", label = "numpy array **2")
new_axes.plot(numpy_array1, numpy_array1 **3, label = "numpy array **3")
new_axes.legend() # we can add location : loc=1,2,3...
plt.show()
```
