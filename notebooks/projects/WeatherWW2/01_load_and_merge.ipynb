```python
df_weather.drop(["WindGustSpd", "PoorWeather", "DR", "SPD", "SND","FT", "FB", "FTI", "ITH", "PGT", "TSHDSBRSGF", "SD3", "RHX", "RHN", "RVG", "WTE" ], axis=1, inplace=True)
```

```python
df_stations.drop(["LAT", "LON"], axis=1, inplace=True)
```

```python
df_stations.info()
```

```python
df_merged = df_weather.merge(df_stations, left_on="STA", right_on="WBAN", how="left")
```

```python
df_merged.drop(columns=["WBAN"], inplace=True)
```
